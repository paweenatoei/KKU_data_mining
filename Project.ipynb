```
path = '/content/drive/My Drive/TB_pro'
```

```
tmdb_5000_movies = os.path.join(path,'tmdb_5000_movies.csv')
```

```
tmdb_5000_movies_df = pd.read_csv(tmdb_5000_movies)
tmdb_5000_movies_df
```

```
tmdb_5000_movies_df.isnull().any()
```

```
tmdb_5000_credits = os.path.join(path,'tmdb_5000_credits.csv')
```

```
tmdb_5000_credits_df = pd.read_csv(tmdb_5000_credits)
tmdb_5000_credits_df
```

```
tmdb_5000_credits_df.isnull().any()
```

```
merge_tb = tmdb_5000_movies_df.merge(tmdb_5000_credits_df,how='left', left_on='movie_id',right_on='movie_id')
merge_tb
```

```
table = merge_tb.groupby('movie_id')['original_title'].apply(list)
table
```
