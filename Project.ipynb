```
merge_tb_dp = merge_tb.dropna()
merge_tb_dp
#เป็นการลบข้อมูลที่ missing ออก
```

```
merge_tb_dp.isnull().any()
#หลังจากที่ทำการลบ missing ออกแล้ว จะทำการตรวจสอบว่าจากที่ทำการลบไปยังมีข้อมูลที่ missing อยู่ไหม
#false คือ ไม่มีข้อมูล missing
#true คือ มีข้อมูลที่ missing
#จากการตรวจสอบพบว่าไม่มีข้อมูลที่ missing แล้ว
```

#Percent of missing data from dropna

##เป็นการตรวจสอบว่ามีข้อมูล missing กี่ % ของข้อมูลทั้งหมด

```
size_before_drop = 4803
size_after_drop = 1493
print(f'size before drop = {size_before_drop} \nsize after drop = {size_after_drop}\nwhich is {100*((size_before_drop-size_after_drop)/size_before_drop)}')
#จากการตรวจสอบพบว่ามีข้อมูล missing อยู่ 68.92%
```

#Data Mining


##จากนั้นจะแบ่งกลุ่มภาพยนต์ออกเป็น 2 กลุ่ม
###กลุ่มที่ 1 คือกลุ่มภาพยนต์ที่มีคะแนนน้อยกว่าหรือเท่ากับ 5
###กลุ่มที่ 2 คือกลุ่มภาพยนต์ที่มีคะแนนมากว่า 5

```
table2 = merge_tb_dp[merge_tb_dp['vote_average']<= 5]
table2
#เป็นการแสดงภาพยนค์ที่มีคะแนนโหวตน้อยว่าหรือเท่ากับ 5
```

```
table3 = merge_tb_dp[merge_tb_dp['vote_average']> 5]
table3
#เป็นการแสดงภาพยนค์ที่มีคะแนนโหวตมากกว่า 5
```

###จากนั้นจะเป็นการแสดงเฉพาะข้อมูลที่เราต้องการนำไปหาความสัมพันธ์

```
table7 = table2.groupby('vote_average')['cast_new'].apply(list)
table7
#เป็นการแสดงข้อมูลของภาพยนต์ที่มีคะแนนน้อยกว่าหรือเท่ากับ 5 ว่ามีนักแสดงชื่ออะไรบ้าง
```

```
table6 = table3.groupby('vote_average')['cast_new'].apply(list)
table6
#เป็นการแสดงข้อมูลของภาพยนต์ที่มีคะแนนมากกว่า 5 ว่ามีนักแสดงชื่ออะไรบ้าง
```

```
!pip install apyori
#เป็นการติดตั้ง apyori เพื่อใช้ในการหาความสัมพันธ์
```

```
from apyori import apriori
```

```
watch = list(apriori(table6, min_support = 0.042 , min_confidence = 0.05))
watch 
#เป็นหาความสัมพันธ์ของนักแสดงกับกลุ่มภาพยนต์ที่คะแนนมากกว่า 5
```

#Summary
###จากการทำ Association Rule จะเห็นได้ว่า
ภาพยนตร์ที่มีคะแนนมากกว่า 5 จากคะแนนเต็ม 10

ที่มีรายชื่อนักแสดงดังนี้

> -"Elijah Wood", "Ian McKellen", "Viggo Mortensen", "Liv Tyler", "Orlando Bloom" มีผลต่อคะแนนของภาพยนต์ 5.88%

> -"Jennifer Lawrence", "Josh Hutcherson", "Liam Hemsworth", "Woody Harrelson", "Elizabeth Banks" มีผลต่อคะแนนของภาพยนต์ 8.82%

> สรุปได้ว่า นักแสดงไม่มีความสัมพันธ์กับคะแนนของภาพยนต์ หรือไม่มีผลต่อคะแนนของภาพยนต์ โดยนักแสดงไม่ได้ทำให้คะแนนของภาพยนต์มากขึ้น หรือน้อยลง แต่คะแนนของภาพยนต์ที่มากหรือน้อยลง อาจขึ้นอยู่กับคุณภาพ และเนื้อเรื่องของภาพยนต์ว่ามีคุณภาพที่ดี หรือเนื้อเรื่องที่ดี และน่าสนใจหรือไม่
